1. Write a function which count number of lines and number of words in a text. All the files are in the data the folder: a) Read obama_speech.txt file and count number of lines and words b) Read michelle_obama_speech.txt file and count number of lines and words c) Read donald_speech.txt file and count number of lines and words d) Read melina_trump_speech.txt file and count number of lines and words

```python
import os

def count_lines_and_words(file_path):
    """
    This counts the number of lines and words in a given text file.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            line_count = len(lines)
            word_count = sum(len(line.split()) for line in lines)
        return line_count, word_count
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return 0, 0

# Folder containing the text files
data_folder = "data"

# List of file names
files = [
    "obama_speech.txt",
    "michelle_obama_speech.txt",
    "donald_speech.txt",
    "melina_trump_speech.txt"
]

# Iterating over each file and counting lines and words
for file_name in files:
    file_path = os.path.join(data_folder, file_name)
    lines, words = count_lines_and_words(file_path)
    print(f"{file_name}: {lines} lines, {words} words")
```

```python
#2.

import json
from collections import Counter

def most_spoken_languages(filename, n):
    """
    Finding the most spoken languages in the world based on the countries data.
    """

    try:
        # Load the JSON data
        with open(filename, 'r', encoding='utf-8') as file:
            countries = json.load(file)
        
        # Create a list of all spoken languages
        all_languages = []
        for country in countries:
            all_languages.extend(country.get('languages', []))
        
        # Count occurrences of each language
        language_counts = Counter(all_languages)
        
        # Get the top N most spoken languages
        most_spoken = language_counts.most_common(n)
        return most_spoken
    
    except FileNotFoundError:
        print(f"File not found: {filename}")
        return []
    except json.JSONDecodeError:
        print("Error decoding JSON. Please ensure the file is valid.")
        return []


filename = './data/countries_data.json'

# The Top 10 most spoken languages
print(most_spoken_languages(filename, 10))

# The Top 3 most spoken languages
print(most_spoken_languages(filename, 3))
```

```python
#3. import json

def most_populated_countries(filename, n):
    """
    Finding the most populated countries in the world based on the countries data.
   """
    try:
        # Load the JSON data
        with open(filename, 'r', encoding='utf-8') as file:
            countries = json.load(file)
        
        # Extract country and population information
        country_population = [
            {"country": country["name"], "population": country["population"]}
            for country in countries
        ]
        

        country_population = sorted(country_population, key=lambda x: x["population"], reverse=True)
        

        return country_population[:n]
    
    except FileNotFoundError:
        print(f"File not found: {filename}")
        return []
    except json.JSONDecodeError:
        print("Error decoding JSON. Please ensure the file is valid.")
        return []


filename = './data/countries_data.json'

# The Top 10 most populated countries
print(most_populated_countries(filename, 10))

# The Top 3 most populated countries
print(most_populated_countries(filename, 3))
```
